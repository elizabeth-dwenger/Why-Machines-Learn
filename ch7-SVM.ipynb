# SVM

### **1. What is SVM?**
Support Vector Machine (SVM) is a **supervised learning algorithm** used for **binary classification**. It finds an **optimal hyperplane** that separates two classes with the **maximum margin**.

Given a dataset $(X_i, y_i)$ where $y_i \in \{-1,1\}$, SVM finds the decision function:

$$
f(x) = w^T x + b
$$

where:
- $w$ is the weight vector.
- $b$ is the bias term.
- $x$ is the input feature vector.

---

### **2. Optimization Problem (Hard Margin SVM)**

The goal is to **maximize the margin** between the two classes:

$$
\min_{w, b} \frac{1}{2} ||w||^2
$$

subject to:

$$
y_i (w^T x_i + b) \geq 1, \quad \forall i
$$

If the dataset is **not linearly separable**, **slack variables**  $\xi_i$ are introduced for soft-margin SVM:

$$
\min_{w, b, \xi} \frac{1}{2} ||w||^2 + C \sum_{i=1}^{n} \xi_i
$$

subject to:

$$
y_i (w^T x_i + b) \geq 1 - \xi_i, \quad \xi_i \geq 0
$$

where C is a **regularization parameter**.

---

### **3. Dual Form and Kernel Trick**
The **dual formulation** uses **Lagrange multipliers** $\alpha$:

$$
\max_{\alpha} \sum_{i=1}^{n} \alpha_i - \frac{1}{2} \sum_{i,j} \alpha_i \alpha_j y_i y_j K(x_i, x_j)
$$

subject to:

$$
\sum_{i=1}^{n} \alpha_i y_i = 0, \quad 0 \leq \alpha_i \leq C
$$

where $K(x_i, x_j)$ is the **kernel function**, allowing **non-linear classification**:

- **Linear Kernel**: $K(x, z) = x^T z$
- **Polynomial Kernel**: $K(x, z) = (1 + x^T z)^d$
- **RBF Kernel**: $K(x, z) = \exp\left(-\frac{\|x - z\|^2}{2\sigma^2}\right)$

---

### **4. Predictions**
Given a new point x, the decision function is:

$$
f(x) = \sum_{i=1}^{n} \alpha_i y_i K(x_i, x) + b
$$

where only **support vectors** contribute to classification.

```python
import numpy as np
import cvxopt
import matplotlib.pyplot as plt

class SVM:
    def __init__(self, kernel='linear', C=None, sigma=1.0):
        self.kernel = kernel
        self.C = C  # Regularization parameter
        self.sigma = sigma  # Used for RBF kernel
        self.alphas = None
        self.b = 0
        self.support_vectors = None
        self.support_vector_labels = None
        self.weights = None

    def linear_kernel(self, x1, x2):
        return np.dot(x1, x2)

    def polynomial_kernel(self, x1, x2, degree=3):
        return (1 + np.dot(x1, x2)) ** degree

    def rbf_kernel(self, x1, x2):
        return np.exp(-np.linalg.norm(x1 - x2) ** 2 / (2 * (self.sigma ** 2)))

    def compute_kernel(self, X):
        n_samples = X.shape[0]
        K = np.zeros((n_samples, n_samples))
        for i in range(n_samples):
            for j in range(n_samples):
                if self.kernel == 'linear':
                    K[i, j] = self.linear_kernel(X[i], X[j])
                elif self.kernel == 'polynomial':
                    K[i, j] = self.polynomial_kernel(X[i], X[j])
                elif self.kernel == 'rbf':
                    K[i, j] = self.rbf_kernel(X[i], X[j])
        return K

    def fit(self, X, y):
        n_samples = X.shape[0]
        K = self.compute_kernel(X)

        # Setup the QP problem
        P = cvxopt.matrix(np.outer(y, y) * K)
        q = cvxopt.matrix(-np.ones(n_samples))
        A = cvxopt.matrix(y, (1, n_samples), tc='d')
        b = cvxopt.matrix(0.0)
        
        if self.C is None:
            G = cvxopt.matrix(-np.eye(n_samples))
            h = cvxopt.matrix(np.zeros(n_samples))
        else:
            G = cvxopt.matrix(np.vstack((-np.eye(n_samples), np.eye(n_samples))))
            h = cvxopt.matrix(np.hstack((np.zeros(n_samples), np.full(n_samples, self.C))))

        # Solve the QP problem
        solution = cvxopt.solvers.qp(P, q, G, h, A, b)
        alphas = np.ravel(solution['x'])

        # Extract support vectors
        support_vector_indices = alphas > 1e-5
        self.alphas = alphas[support_vector_indices]
        self.support_vectors = X[support_vector_indices]
        self.support_vector_labels = y[support_vector_indices]

        # bias term
        self.b = np.mean([y_k - np.sum(self.alphas * self.support_vector_labels * K[idx][support_vector_indices])
                          for idx, y_k in enumerate(y[support_vector_indices])])

        # weight vector (linear SVM)
        if self.kernel == 'linear':
            self.weights = np.sum(self.alphas[:, np.newaxis] * self.support_vector_labels[:, np.newaxis] * self.support_vectors, axis=0)

    def predict(self, X):
        if self.kernel == 'linear':
            return np.sign(np.dot(X, self.weights) + self.b)
        else:
            y_pred = np.zeros(X.shape[0])
            for i in range(X.shape[0]):
                sum_alpha_y_k = 0
                for alpha, sv_y, sv in zip(self.alphas, self.support_vector_labels, self.support_vectors):
                    sum_alpha_y_k += alpha * sv_y * self.rbf_kernel(X[i], sv)
                y_pred[i] = sum_alpha_y_k + self.b
            return np.sign(y_pred)
```

```python
def plot_decision_boundary(svm, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

    Z = np.array([svm.predict(np.array([[xx_val, yy_val]])) for xx_val, yy_val in zip(xx.ravel(), yy.ravel())])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6), dpi=300)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    
    # Plot data 
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k', s=100)
    
    # Highlight support vectors
    plt.scatter(svm.support_vectors[:, 0], svm.support_vectors[:, 1], s=200, edgecolors='k', facecolors='none', linewidths=2, label='Support Vectors')

    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title(f"SVM Decision Boundary ({svm.kernel} kernel)")
    plt.legend()
    plt.show()
```

```python
if __name__ == "__main__":
    X = np.array([[2, 3], [1, 1], [2, 1], [3, 3], [2, 2], [3, 2]])
    y = np.array([1, -1, -1, 1, -1, 1])

    svm = SVM(kernel='rbf', C=1.0, sigma=1.0)
    svm.fit(X, y)

    plot_decision_boundary(svm, X, y)
```
